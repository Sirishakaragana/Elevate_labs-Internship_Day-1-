```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('/content/Netflix_movies_and_tv_shows_clustering.csv.zip')
```

```python
df.head()
```

```python
# Clean column names
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
```

```python
# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.isnull().values.any()
```

```python
df.isnull().sum().sum()
```

```python
#Null values are present in each column
df.isnull().sum()
```

```python
# Fill missing values
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Provided')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].ffill()
```

```python
df.isnull().sum()
```

```python
df['rating'] = df['rating'].fillna('Not Rated')


```

```python
df = df.dropna(subset=['rating'])
```

```python
df.isnull().sum()
df.isnull().values.any()
# Dataset is now clean — all null values have been handled
```

```python
# Standardize text fields
text_columns = ['type', 'title', 'director', 'cast', 'country', 'rating', 'duration', 'listed_in', 'description']
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df[text_columns].head()
#text columns are updated
```

```python
# Export cleaned dataset
df.to_csv('cleaned_netflix_data.csv', index=False)
```
